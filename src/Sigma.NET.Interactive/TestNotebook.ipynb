```polyglot-notebook
open Sigma.NET
open Sigma.NET.Interactive

let rnd = new Random()
let colors = [|"#F2FFE9";"#A6CF98";"#557C55";"#FA7070"|]

VisGraph.empty()
|> VisGraph.withNodes [
        node "n1" |> withNodeData(Label="AAA", X=1, Y=2, Size=rnd.Next(8, 18), Color=colors.[rnd.Next(0, 4)])
        node "n2" |> withNodeData(Label="BBB", X=3, Y=4, Size=rnd.Next(8, 18), Color=colors.[rnd.Next(0, 4)])
        node "n3" |> withNodeData(Label="CCC", X=5, Y=8, Size=rnd.Next(8, 18), Color=colors.[rnd.Next(0, 4)])
    ]
|> VisGraph.withEdges [
        edge "n1" "n2"
        edge "n2" "n3"
    ]
|> VisGraph.withCircularLayout()
|> VisGraph.withRenderer(Render.Settings.Init(LabelColor=Render.ColorOrReference.Init(Color="#FA7070"))) 
|> VisGraph.withSize(CssLength.Percent(100), CssLength.PX(300))
```

```polyglot-notebook

open Graphoscope
open Graphoscope.RandomModels

let colors = [|"#F2FFE9";"#A6CF98";"#557C55";"#FA7070"|]

let N = 1000
let p = 0.005
let rnd = new Random()
let random_number = rnd.Next(1, 51)

let myBollobasRiordan = RandomModels.BollobasRiordan.initDirectedFGraph  N 0.5 0.1 0.4 0.6 0.4 FGraph.empty

VisGraph.empty()
|> VisGraph.withNodes[
    for node in myBollobasRiordan do
        yield (Node.Init(key=string node.Key,DisplayData=DisplayData.Init(Size=rnd.Next(1, 18), Color=colors.[rnd.Next(0, 4)], Label=string node.Key)))

]
|> VisGraph.withEdges [
    for node in myBollobasRiordan do
        let pred,t,_ = node.Value
        for kv in pred do
            yield (Edge.Init(source=string kv.Key, target=string t,DisplayData=DisplayData.Init(Size=rnd.Next(1, 3))) )

]

|> VisGraph.withRenderer(Render.Settings.Init(LabelColor=Render.ColorOrReference.Init(Color="#FA7070"))) 
|> VisGraph.withForceAtlas2(Iterations=500,Settings=FA2Settings.Init(AdjustSizes=true,Gravity=2),GetEdgeWeight="size")
// |> VisGraph.withNoverlap(50)
//|> VisGraph.withCircularLayout()
```
